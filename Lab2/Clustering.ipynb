# Task 2.1 Fisher Iris Dataset

# Loading Dataset
Iris contains 150 instances, multivariate, with 4 attributes: 
1. sepal length in cm
2. sepal width in cm
3. petal length in cm
4. petal width in cm


Iris Labels contains numbers {0, 1, 2}, note each data point as being one of the following flowers 
-- Iris Setosa
-- Iris Versicolour
-- Iris Virginica

```python
data = np.load('Iris_data.npy')
labels = np.load('Iris_labels.npy')
```

```python
print(data.shape)
```

```python
print(labels)
```

```python
plt.scatter(data[:,0], data[:,1], c=labels)

# Add title and axis names
plt.title('Iris Data according to Labels')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')


plt.show()
```

# Task 2.2 k-Means

cluster our data with k-means using unsupervised learning.
Cluster using all feature dimensions of Iris dataset into k clusters

```python
myKmeansModel = sklearn.cluster.KMeans()
```

```python
myKmeansModel = myKmeansModel.fit(data)
```

```python
predictions = myKmeansModel.predict(data)
print(predictions.shape)
```

```python
print(predictions)
```

```python
plt.scatter(data[:,0], data[:,1], c=predictions)

# Add title and axis names
plt.title('Iris Data Predictions')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')

plt.show()
```

# Adding centroids and attempt clustering into the 3 labels and adding centroid plots

```python
myKmeansModel1 = sklearn.cluster.KMeans(n_clusters=3)

myKmeansModel1 = myKmeansModel1.fit(data)

centroids = myKmeansModel1.cluster_centers_

predictions = myKmeansModel1.predict(data)


#make graph

fig = plt.figure()
ax1 = fig.add_subplot(111)

ax1.scatter(data[:,0], data[:,1], c=predictions)
ax1.scatter(centroids[:,0], centroids[:,1], c='red', label="centroids")
plt.legend(loc='upper right');

# Add title and axis names
plt.title('Iris Data Predictions')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')

plt.show()
```
