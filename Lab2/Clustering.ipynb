```python
import numpy as np
import matplotlib.pyplot as plt
import sklearn.cluster
```

# Task 2.1 Fisher Iris Dataset

# Loading Dataset
Iris contains 150 instances, multivariate, with 4 attributes: 
1. sepal length in cm
2. sepal width in cm
3. petal length in cm
4. petal width in cm


Iris Labels contains numbers {0, 1, 2}, note each data point as being one of the following flowers 
-- Iris Setosa
-- Iris Versicolour
-- Iris Virginica

```python
data = np.load('Iris_data.npy')
labels = np.load('Iris_labels.npy')
```

```python
print(data.shape)
```

```python
print(labels)
```

```python
plt.scatter(data[:,0], data[:,1], c=labels)

# Add title and axis names
plt.title('Iris Data according to Labels')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')


plt.show()
```

# Task 2.2 k-Means

```python
myKmeansModel = sklearn.cluster.KMeans()
```

```python
myKmeansModel = myKmeansModel.fit(data)
```

```python
predictions = myKmeansModel.predict(data)
print(predictions.shape)
```
